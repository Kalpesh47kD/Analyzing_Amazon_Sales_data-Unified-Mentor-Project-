```python
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_csv('/content/Amazon Sales data.csv')

df.head(20)
```

```python
df['Order Date'] = df['Order Date'].str.replace('-', '/')
df['Ship Date'] = df['Ship Date'].str.replace('-', '/')
```

```python
df.head(20)
df = df.astype(str)
columns = df.columns.tolist()
print(columns)
for column in columns:
    df[column] = df[column].str.strip()
```

```python
df.head(20)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df['Order Date'] = df['Order Date'].astype(str)
df['Year'] = df['Order Date'].apply(lambda x: x.split('-')[0])
df['Month'] = df['Order Date'].apply(lambda x: x.split('-')[1])

df.head()
```

```python
for column in columns[8:]:
    df[column] = df[column].astype(float)

    df.head()
```

```python
df = df.drop(['Order ID'],axis=1)
df.head()
```

```python
df = df.drop(['Order Date'],axis=1)

df.head()
```

```python
import plotly.express as px
fig = px.pie(df,values="Total Revenue",names="Item Type",template="plotly_dark",title="Total revenue per item")
fig.show()
```

```python
fig = px.pie(df,values="Total Revenue",names="Region",template="plotly_dark",title="Total revenue in Regions")
fig.show()
```

```python
fig = px.scatter(df,x='Country',y='Units Sold',color='Region',size='Units Sold',animation_frame='Year',template='plotly_dark',title="Units Sold Per Year")
fig.show()
```

```python
fig = px.scatter(df,x='Country',y='Units Sold',color='Region',size='Units Sold',animation_frame='Month',template='plotly_dark',title='Units Sold Per Month')
fig.show()
```

```python
fig = px.bar(df,x='Country',y=["Total Cost", "Total Profit","Total Revenue"],barmode='stack',animation_frame='Year',template="plotly_dark",title="combination of all factors")
fig.show()
```

```python
fig = px.bar(df,x='Country',y=["Total Cost", "Total Profit","Total Revenue"],barmode='stack',animation_frame='Month',template="plotly_dark",title="combination of all factors per month")
fig.show()
```

```python
df['Net Profit Margin'] = (df['Total Profit'] * 100)/df['Total Revenue']
df.head()
```

```python
fig=px.scatter(df,x='Country',y='Net Profit Margin',color='Region',animation_frame='Year',size='Net Profit Margin',template='plotly_dark')
fig.show()
```

```python
fig=px.scatter(df,x='Country',y='Net Profit Margin',color='Region',animation_frame='Month',size='Net Profit Margin',template='plotly_dark',title='Net profit Margin Per month')
fig.show()
```

```python
fig = px.scatter(df,x='Country',y='Units Sold',color='Item Type',size='Units Sold',animation_frame='Year',template='plotly_dark',title="Units Sold Per Item")
fig.show()
```
